```python
from sklearn.model_selection import train_test_split

X_train, X_holdout, y_train, y_holdout = train_test_split(
    X_data, y_data, test_size=0.3, random_state=0
)
```

Exercise: network architecture
- input dimension: number of features (measures in Basel? or everywhere?)
- no activation
- two outputs: precipitation and sunshine

```python
from tensorflow import keras 

def create_nn(df):
    n_features = df.shape[1]
    inputs = keras.Input(shape=(n_features,), name="input")
    
    layers_dense = keras.layers.Dense(100, "relu")(inputs)
    layers_dense = keras.layers.Dense(50, "relu")(layers_dense)
    
    outputs = keras.layers.Dense(1)(layers_dense) # 1 node, no activation
    
    return keras.Model(inputs=inputs, outputs=outputs, name="weather_prediction_model")
```

```python
model = create_nn(X_data)
```

```python
model.summary() # number of params: (n_neurons current) x (n_neurons previous + 1), 1 for the bias term
```

```python
model.compile(loss="mse")
```

```python
model.compile(optimizer="adam", loss="mse")
```

we can add more metrics to monitor how the learning is going: `metrics` argument in `model.compile`

```python
model.compile(optimizer="adam", loss="mse", metrics=[keras.metrics.RootMeanSquaredError()])
```

```python
def compile_model(model):
    model.compile(optimizer='adam',
                  loss='mse',
                  metrics=[keras.metrics.RootMeanSquaredError()])
```

```python
compile_model(model)
```

trade-off for batch sizes: larger batches = better gradients, but parameters are updated less often

```python
history = model.fit(
    X_train, y_train, batch_size=32, epochs=200, verbose=2
)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

def plot_history(metrics):
    """
    Plot the training history

    Args:
        metrics(str, list): Metric or a list of metrics to plot
    """
    history_df = pd.DataFrame.from_dict(history.history)
    sns.lineplot(data=history_df[metrics])
    plt.xlabel("epochs")
    plt.ylabel("error")
    

```

```python
plot_history(["root_mean_squared_error", "loss"])
```

```python
y_train_predicted = model.predict(X_train)
y_test_predicted = model.predict(X_test)
```

```python
def plot_predictions(y_pred, y_true, title):
    plt.style.use("ggplot")
    plt.scatter(y_pred, y_true, s=10, alpha=0.5)
    plt.xlabel("predicted sunshine hours")
    plt.ylabel("true sunshine hours")
    plt.title(title)
    
```

```python
plot_predictions(y_train_predicted, y_train, title="predictions on training")
```

```python
plot_predictions(y_test_predicted, y_test, title='Predictions on the test set')
```

```python
train_metrics = model.evaluate(X_train, y_train, return_dict=True)
test_metrics = model.evaluate(X_test, y_test, return_dict=True)
```

```python
print('Train RMSE: {:.2f}, Test RMSE: {:.2f}'.format(train_metrics['root_mean_squared_error'], test_metrics['root_mean_squared_error']))
```

```python
y_baseline_prediction = X_test["BASEL_sunshine"]
```

Exercise: plot the training process
- difference between training and validation data: overfitting -- the validation set error increases, while the training set error decreases 
- ideally: curves should overlap until the end of training.

```python

def create_flexible_nn(df, nodes=(100, 50)):
    n_features = df.shape[1]
    inputs = keras.Input(shape=(n_features,), name="input")
    
    layers_dense = keras.layers.Dense(nodes[0], "relu")(inputs)
    layers_dense = keras.layers.Dense(nodes[1], "relu")(layers_dense)
    
    outputs = keras.layers.Dense(1)(layers_dense) # 1 node, no activation
    
    return keras.Model(inputs=inputs, outputs=outputs, name="flexible_weather_prediction_model")
                       
                 
```

```python
node_combinations = [
    (100, 50), 
    (10, 50),
    (100, 5),
    (10, 5)
]

import numpy as np 
results = {}
for combination in node_combinations:
    model = create_flexible_nn(X_data, combination)
    compile_model(model)
    history = model.fit(
        np.array(X_train), np.array(y_train), epochs=200, validation_data=(X_val, y_val)
    )
    results[combination] = history
```

```python
def plot_history(history, metrics):
    """
    Plot the training history

    Args:
        metrics(str, list): Metric or a list of metrics to plot
    """
    history_df = pd.DataFrame.from_dict(history.history)
    sns.lineplot(data=history_df[metrics])
    plt.xlabel("epochs")
    plt.ylabel("error")
```

```python
for combination, h in results.items():
    plot_history(h, ["root_mean_squared_error", "val_root_mean_squared_error"])
```

```python
def create_nn():
    # Input layer
    inputs = keras.layers.Input(shape=(X_data.shape[1],), name='input')

    # Dense layers
    layers_dense = keras.layers.BatchNormalization()(inputs) # we add this layer, relative to the previous model
    layers_dense = keras.layers.Dense(100, 'relu')(layers_dense)
    layers_dense = keras.layers.Dense(50, 'relu')(layers_dense)

    # Output layer
    outputs = keras.layers.Dense(1)(layers_dense)

    # Defining the model and compiling it
    return keras.Model(inputs=inputs, outputs=outputs, name="model_batchnorm")
    
```

### batch normalization 

documentation: [link](https://keras.io/api/layers/normalization_layers/batch_normalization/)
- during training, normalization is based on the current value of the current batch of inputs
- during inference, inputs are only normalized if the inference data are similar to the training data
    - what does "similar" mean here?
    - I think the `momentum` parameter, which is used to update the moving averages/variances, stems only from the training? ie, after training, self.moving_var and self.moving_mean are fixed?
    - remember from Goodfellow et al 2016: momentum stabilizes the gradient and the inverse of the Hessian (ill-conditioning)
- I suppose when the model is run on the validation set (but during training), then the inference normalization applies? 

why do we do batch normalization? 
- same reason as for conventional ML? to have all features on same scale, so that differently scaled variables do not get selected more/less often (in a regularization model?)
- see Goodfellow et al, 2016, ch. 8.7.1: the gradient defines how much one parameter should be updated, holding the others fixed. But all the others change simultaneously! This renders the gradient unstable (zig-zagging?)--particularly for large models--, and perhaps also the Hessian (intuition?). So, batch normalization induces some persistence in the parameter updates so that the update in the parameters at $t$ depends not only on the current gradient but also on the (decaying) history of the past gradients $t-k$.
